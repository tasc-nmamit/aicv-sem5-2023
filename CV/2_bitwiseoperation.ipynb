```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Create a function to make things easy
def display(texts, img):
    cv2.imshow(texts, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img1 = cv2.imread('../imagedata/1bit1.png')
img2 = cv2.imread('../imagedata/1bit2.png')
```

```python
# Image Opeation AND
bitwiseAnd = cv2.bitwise_and(img1, img2, mask=None) #It creates a region on which the operation should be applied.
display("AND", bitwiseAnd)
```

```python
# Image Operation OR
bitwiseOr = cv2.bitwise_or(img1, img2)
display("OR", bitwiseOr)
```

```python
# Image Operation Not
bitswiseNot = cv2.bitwise_not(img2)
display("NOT", bitswiseNot)
```

```python
# Image Operation XOR
bitwiseXor = cv2.bitwise_xor(img1,img2)
display("XOR", bitwiseXor)
```
