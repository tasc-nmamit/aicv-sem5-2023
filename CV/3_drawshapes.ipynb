A simple program to draw geometric shapes

```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
img = np.zeros((500,500,3),dtype="uint8")
img[0:500,:]= [138,138,138]
```

```python
plt.imshow(img)
plt.show()
```

```python
# Draw Line
cv2.line(img, (10,100),(100,100),(0,0,255),10)
# Draw Rectangle
cv2.rectangle(img, (100,10),(200,200),(1,2,123),5)
# Draw Circle
cv2.circle(img, (150,250),45,(66,224,245),5)
# Displaying Text
cv2.putText(img, "Hampter",(300,200),cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,0),2, cv2.LINE_AA)


plt.imshow(img)
plt.show()
```
